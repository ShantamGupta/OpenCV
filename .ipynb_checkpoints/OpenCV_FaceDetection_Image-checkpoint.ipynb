## Load the libraries

```python
import cv2
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = os.getcwd()
```

## Read an Image

```python
img = cv2.imread(path + '\\Image.jpg',1)
plt.imshow(img)
```

## Load the Haar Cascade filters from OpenCV library

```python
cv2.startWindowThread() 
face_cascade = cv2.CascadeClassifier(path + '\\haarcascades\\haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(path + '\\haarcascades\\haarcascade_eye.xml')
img = cv2.imread(path + '\\Image.jpg',1)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

## Display Rectangles

```python
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x,y,w,h) in faces:
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
